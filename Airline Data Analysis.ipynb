# How many number of ticket booked and total amount earned change with time.

```python
ticket=pd.read_sql_query("""select * from tickets inner join bookings on tickets.book_ref=bookings.book_ref""",connection)
ticket['book_date']=pd.to_datetime(ticket['book_date'])
ticket['date']=ticket['book_date'].dt.date
x=ticket.groupby('date')[['date']].count()
plt.figure(figsize=(18,6))
plt.plot(x.index,x['date'],marker ='^')
plt.xlabel('Date',fontsize=20)
plt.ylabel('Number of tickets',fontsize=20)
plt.grid('b')
plt.show()
```

Here's a summary of the observation regarding ticket bookings:

1.June 22nd to July 7th: Gradual increase in the number of tickets booked.
2.July 8th to August 8th: Relatively stable ticket bookings with a noticeable peak on a single day.
3.August 9th to August 14th: Increase in ticket bookings, approximately up to 10,000, likely due to a   seasonal effect.
4.August 15th: Decrease in ticket bookings.

```python
bookings=pd.read_sql_query("""select * from bookings;""",connection)
bookings['book_date']=pd.to_datetime(bookings['book_date'])
bookings['date']=bookings['book_date'].dt.date
x=bookings.groupby('date')[['total_amount']].sum()
plt.figure(figsize= (18,6))
plt.plot(x.index,x['total_amount'],marker = '^')
plt.xlabel('Date',fontsize = 20)
plt.ylabel('Total amount earned', fontsize= 20)
plt.grid('b')
plt.show()
```

Observation
1.The Total amount earned by the company is closely correlated with the number of tickets booked.
2.The total revenue earned followed a similar trend throughout the analyzed time period.
3.Further exploration of the factors contributing to the peak in ticket bookings could help increase overall revenue and optimize operational strategies.

Calculate the average occupancy per aircraft

```python
occupancy_rate=pd.read_sql_query("""select a.aircraft_code,avg(a.seats_count) as booked_seats, b.num_seats, avg(a.seats_count)/b.num_seats as occupancy_rate
                     from
                     (select aircraft_code,flights.flight_id,count(*) as seats_count from boarding_passes inner join flights
                     on boarding_passes.flight_id=flights.flight_id group by aircraft_code, flights.flight_id)as a inner join
                    (select aircraft_code,count(*) as num_seats from seats group by aircraft_code) as b on a.aircraft_code=b.aircraft_code
                     group by a.aircraft_code""", connection)
occupancy_rate
```

calculate by how much the total annual turnover could increase by givivng all aircraft a 10% higer occupancy rates.

```python
occupancy_rate['Inc occupancy rates']= occupancy_rate['occupancy_rate']+ occupancy_rate['occupancy_rate']*0.1
occupancy_rate
```

```python
total_revenue= pd.read_sql_query("""select aircraft_code, sum(amount)as total_revenue from ticket_flights join flights on ticket_flights.flight_id = flights.flight_id group by aircraft_code""",connection)
total_revenue
```

```python
occupancy_rate['Inc total annual turnover']= total_revenue['total_revenue']/occupancy_rate['occupancy_rate']*occupancy_rate['Inc occupancy rates']
occupancy_rate
```

Observation:
The average occupancy per ticket is another critical number to consider. Airlines may measure how sucessfully they fill their seats and discover chances to boost occupancy rates by using their metrics. Heigher occupancy rates can help airlines increase revenue and profitability while lowering operational expenses associated with vacant seats.  
Pricing strategy, airline shedule and customer satisfaction are all factors that might influence occupancy rates. Heigher occupany rates means the aircraft seats are more booked and only few seats are left unbooked. The total revenue increased after increasing the occupancy rates by 10% and its gives the result that it will increase gradually so airlines should be more focused on the pricing strategies.
